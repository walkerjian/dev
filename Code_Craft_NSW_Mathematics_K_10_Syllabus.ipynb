```python
import graphviz

# Create a new directed graph for the flowchart
dot = graphviz.Digraph(comment='NSW Mathematics K-10 Syllabus')

# Add nodes and edges to represent the strands and substrands
dot.node('A', 'NSW Mathematics K-10 Syllabus')
dot.node('B', 'Number and Algebra')
dot.node('C', 'Measurement and Geometry')
dot.node('D', 'Statistics and Probability')

dot.node('E', 'Number and place value')
dot.node('F', 'Patterns and algebra')
dot.node('G', 'Fractions and decimals')
dot.node('H', 'Money and financial mathematics')

dot.node('I', 'Using units of measurement')
dot.node('J', 'Shape')
dot.node('K', 'Geometric reasoning')
dot.node('L', 'Location and transformation')

dot.node('M', 'Chance')
dot.node('N', 'Data representation and interpretation')

dot.edges(['AB', 'AC', 'AD'])
dot.edges(['BE', 'BF', 'BG', 'BH'])
dot.edges(['CI', 'CJ', 'CK', 'CL'])
dot.edges(['DM', 'DN'])

# Render the graph to a file and display it
dot.render('/mnt/data/Mathematics_Syllabus_Flowchart', format='png', cleanup=True)
dot

```
