```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.0, 1, 300)
y = np.sin(2*np.pi*x)

# Round the sine wave values
y_rounded = np.round(y)

plt.plot(x, y, label="Sine Wave")
plt.plot(x, y_rounded, label="Rounded Sine Wave", linestyle="--")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Worth A Trillion Dollars?')
#plt.legend()

plt.grid(True)
plt.show()

```
