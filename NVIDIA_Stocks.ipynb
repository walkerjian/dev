```python
from datetime import datetime
import yfinance as yf
import matplotlib.pyplot as plt

# NVIDIA Corporation ticker symbol
ticker_symbol = "NVDA"

# NVIDIA IPO date
nvidia_ipo_date = "1999-01-22"

# Current date
current_date = datetime.now().strftime('%Y-%m-%d')

# Download historical data for NVIDIA
nvidia_data = yf.download(ticker_symbol, start=nvidia_ipo_date, end=current_date)

# Initial investment in USD
initial_investment = 1

# Price of NVIDIA stock on the first day of trading
first_day_close_price = nvidia_data['Close'].iloc[0]

# Number of shares bought with the initial investment
shares_bought = initial_investment / first_day_close_price

# Current value of the investment
nvidia_data['Investment Value'] = shares_bought * nvidia_data['Close']

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(nvidia_data['Investment Value'], label='Investment Value')
plt.title(f'Value of ${initial_investment} Invested in NVIDIA at IPO')
plt.xlabel('Date')
plt.ylabel('Investment Value in USD')
plt.legend()
plt.grid(True)
plt.yscale('log')  # Log scale to better visualize growth over time
plt.tight_layout()

# Display current investment value
current_investment_value = nvidia_data['Investment Value'][-1]

plt.show(), current_investment_value
```
